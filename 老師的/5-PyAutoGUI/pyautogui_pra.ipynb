```python
import pyautogui
"""
小算盤
擷取數字圖片後自動點擊
"""

# 要擷取數字的大小
num_size = 30
# 數字之間的水平與垂直距離
x_distance = 80
y_distance = 50

# 起始位置 (從數字1的位置開始)
start_location = (30, 410)
for i in range(3):
    pyautogui.screenshot(f"{1+i}.png", region=(start_location[0]+x_distance*i,start_location[1],num_size,num_size))

    pyautogui.screenshot(f"{4+i}.png", region=(start_location[0]+x_distance*i,start_location[1] - y_distance,num_size,num_size))

    pyautogui.screenshot(f"{7+i}.png", region=(start_location[0]+x_distance*i,start_location[1] - y_distance*2,num_size,num_size))

pyautogui.screenshot(f"0.png", region=(start_location[0] + x_distance, start_location[1]+y_distance, num_size,num_size))

```

```python
# 執行點擊動作
def click_num(num):
    location = pyautogui.locateCenterOnScreen(f"{num}.png")
    if location:
        pyautogui.click(location)
        
click_num(5)
click_num(2)
click_num(0)
```
