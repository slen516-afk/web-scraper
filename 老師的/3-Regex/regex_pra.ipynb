```python
import re
text = " ID/姓名： E123456789 / Denji"

# 使用正則表達式提取 ID 和姓名
regex = r"ID/姓名： (\w+) / (\w+)"

# 使用 search / match 方法進行匹配
# match = re.match(regex, text)
match = re.search(regex, text)

print(match)
print(match.group(0))
print(match.group(1))
print(match.group(2))
```

```python
import re
text = """
ID/姓名： 123456789 / Denji
身高(cm)/體重(kg)： 170 / 65
"""
# 使用正則表達式提取 ID 和姓名
regex = r"ID/姓名： (\d+) / (\w+)"

data = re.findall(regex, text)
print(data)
for i in data:
    print(f"ID: {i[0]}")
    print(f"name: {i[1]}")

for id, name in data:
    print(f"ID: {id}")
    print(f"name: {name}")
```

### ⚠️ 要使用 `.`、`(`、`)` 時要使用 `\` 跳脫

```python
import re

text = '住家電話：(02)2311-3831'

# 使用正則表達式提取區碼和電話號碼
pattern = r'住家電話：(\(\d{2}\)\d{4}-\d{4}$)'

match = re.search(pattern, text)

print(match.group(0))
print(match.group(1))
```

```python
import re
text = """
    請用以下方式聯絡：
    住家: (02)2311-3838
    公司: (02)2311-3839#345
    手機: 0912-345-678
"""

pattern = r"\(?\d{2,4}\)?-?\d{3,4}-?\d{3,4}(?:#\d{3})?"
pattern = r"[\(]?\d+.\d+.\d+.\d+."
# 使用正則表達式提取多種電話號碼格式
data = re.findall(pattern, text)
print(data)
```

```python
import re
texts = """<table><tr><td colspan="8">動漫評分榜（Summary）</td></tr><tr><td colspan="2">作品名稱</td><td>S</td><td>A</td><td>B</td><td>C</td><td>D</td><td></td></tr><tr><td>1</td><td>航海王（One Piece）</td><td></td><td></td><td>★</td><td></td><td></td><td></td></tr><tr><td>2</td><td>鏈鋸人（Chainsaw Man）</td><td>★</td><td></td><td></td><td></td><td></td><td></td></tr><tr><td>3</td><td>咒術迴戰（Jujutsu Kaisen）</td><td></td><td>★</td><td></td><td></td><td></td><td></td></tr><tr><td>4</td><td>進擊的巨人（Attack on Titan）</td><td>★</td><td></td><td></td><td></td><td></td><td></td></tr><tr><td>5</td><td>一拳超人（One Punch Man）</td><td></td><td>★</td><td></td><td></td><td></td><td></td></tr><tr><td>6</td><td>間諜家家酒（SPY×FAMILY）</td><td></td><td></td><td>★</td><td></td><td></td><td></td></tr><tr><td>7</td><td>死神 千年血戰篇（BLEACH TYBW）</td><td></td><td></td><td></td><td></td><td>★</td><td></td></tr><tr><td>8</td><td>東京復仇者（Tokyo Revengers）</td><td></td><td></td><td></td><td></td><td>★</td><td></td></tr><tr><td>9</td><td>鬼滅之刃（Demon Slayer）</td><td></td><td></td><td></td><td></td><td>★</td><td></td></tr><tr><td>10</td><td>我的英雄學院（My Hero Academia）</td><td></td><td>★</td><td></td><td></td><td></td><td></td></tr></table>"""

# *? 盡可能的少匹配
pattern = r'<tr>(.*?)</tr>'
matches = re.findall(pattern, texts)

grade_dict = {
    0: "S",
    1: "A",
    2: "B",
    3: "C",
    4: "D"
}

anime_grade = {}
for match in matches[2:]:
    # print(match)
    pattern = r'<td>(.*?)</td>'
    info_list = re.findall(pattern, match)
    print(info_list)

    for index, value in enumerate(info_list[2:]):
        if value == '★':
            # 遇到 ★ 對應 dict 的分數
            anime_grade[info_list[1]] = grade_dict[index]

from pprint import pprint
pprint(anime_grade)
```
