```python
import json
import os 
import requests
from bs4 import BeautifulSoup as bs

folder_name = "line_stickers"

# 建立資料夾
if not os.path.exists(folder_name):
    os.mkdir("line_stickers")
else:
    print(f"已建立過 {folder_name}")

def get_stickers(url):
    r = requests.get(url)
    soup = bs(r.text, 'lxml')

    # 貼圖名稱
    sticker_name = soup.find("p", class_="mdCMN38Item01Ttl").text

    if not os.path.exists(f"line_stickers/{sticker_name}"):
        os.mkdir(f"line_stickers/{sticker_name}")
    else:
        print(f"已建立過 {sticker_name}")

    # 找到貼圖清單
    sticker_list = soup.find("ul", class_="mdCMN09Ul FnStickerList")

    # 找到所有貼圖元素
    sticker_elements = sticker_list.find_all('li')

    # 訪問每個貼圖的元素
    for index, sticker_element in enumerate(sticker_elements, start=1):
        print(sticker_element["data-preview"])
        print(type(sticker_element["data-preview"]))
        data_preview = json.loads(sticker_element["data-preview"])
        static_url = data_preview["staticUrl"]

        image_content = requests.get(static_url).content
        # with open(f"line_stickers/{sticker_name}/line_sticker_{index}.png", mode="wb") as f:
        #     f.write(image_content)


get_stickers("https://store.line.me/stickershop/product/28419/zh-Hant")
```
