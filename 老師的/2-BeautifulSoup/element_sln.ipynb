```python
# 引入套件
from bs4 import BeautifulSoup as bs

# HTML elements
html_doc = """
<html>
<body>
<h1>Welcome</h1>
<p class="title"><b>The Chainsaw Man's anime</b></p>
<a href="http://example.com/Pochita" class="character" id="link1">Pochita</a>
<a href="http://example.com/Power" class="character" id="link2">Power</a>
<a href="http://example.com/Makima" class="character" id="link3">Makima</a>
<p class="anime">They are partners in the story</p>
</body>
</html>
"""

# 使用 Beautifule Soup
soup = bs(html_doc, 'lxml')
```

```python
# 找到所有的 a 標籤的元素
elements = soup.find_all("a")
print(elements)

# 找到所有的 a 標籤，並且 class 為 character 的元素
elements = soup.find_all("a", class_="character")
print(elements)

# 找到所有的 a 標籤，並且 id 為 link2 的元素
elements = soup.find_all("a", id="link2")
print(elements)

# 找到第一個 a 標籤而且 href 為 http://example.com/Pochita 的元素
element = soup.find('a', href="http://example.com/Pochita")
print(elements)

# 找到所有內容為 Pochita 的元素
elements = soup.find_all(string="Pochita")
print(elements)
print(elements[0].parent)  # 透過 string 找到的元素，可以使用 parent 屬性找到其父元素
print(elements[0].parent['href'])  # 取得父元素的 href 屬性值
```

```python
# 找到所有的 p 標籤
p_tags = soup.find_all('p')

for p_tag in p_tags:
    print(p_tag)
    # print(p_tag.find('b'))
    print(p_tag.text)
    print(p_tag['class'])
    print('='*30)


# 找到 class=title 的 p 標籤
p_tag_with_title = soup.find('p', class_='title')
print(p_tag_with_title)
```

```python
# 找到網址為 Power 的 a 標籤
element = soup.find('a', href="http://example.com/Power")
print(element)

# 找到內文為 Power 的 a 標籤
element = soup.find('a', string="Power")

print(element)
print(element['href'])
```

```python
# attribute 多條件篩選 
element = soup.find('a', {'class': 'character', 'id': 'link3'})

print(element)
```

### 使用 CSS select

```python
# 引入套件
from bs4 import BeautifulSoup as bs

# HTML elements
html_doc = """
<html>
<body>
<h1>Welcome</h1>
<p class="title"><b>The Chainsaw Man's anime</b></p>
<a href="http://example.com/Pochita" class="character" id="link1">Pochita</a>
<a href="http://example.com/Power" class="character" id="link2">Power</a>
<a href="http://example.com/Makima" class="character" id="link3">Makima</a>
<p class="anime">They are partners in the story</p>
</body>
</html>
"""
```

```python
# 使用 Beautifule Soup
soup = bs(html_doc, 'lxml')

# 使用 CSS select 找到所有的 a 標籤的元素
elements = soup.select("a")
print(elements)

# 使用 CSS select 找到所有的 a 標籤，並且 class 為 character 的元素
elements = soup.select("a.character")
print(elements)

# 使用 CSS select 找到所有的 a 標籤，並且 id 為 link2 的元素
elements = soup.select("a#link2")
print(elements)

# 使用 CSS select 找到連結為 Pochita 的元素
elements = soup.select("a[href='http://example.com/Pochita']")
print(elements)
```

```python
# 模糊搜尋 href 屬性中包含 Power 的元素
elements = soup.select("a[href*='Po']")
print(elements)

# 模糊搜尋 class 結尾為 ter 的元素
elements = soup.select("a[class$='ter']")
print(elements)

# 模糊搜尋 id 包含 link 的元素
elements = soup.select("a[id*='link']")
print(elements)
```
