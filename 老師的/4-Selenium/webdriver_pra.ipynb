```python
from selenium import webdriver
from selenium.webdriver.common.by import By

# 設定 Chrome 瀏覽器的選項
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized") # Chrome 瀏覽器在啟動時最大化視窗
options.add_argument("--incognito") # 無痕模式
options.add_argument("--disable-popup-blocking") # 停用 Chrome 的彈窗阻擋功能。

# 建立 Chrome 瀏覽器物件
driver = webdriver.Chrome(options=options)
driver.get("https://www.104.com.tw/")
```

```python
job_cards = driver.find_elements(By.CSS_SELECTOR, ".vue-recycle-scroller__item-view.recycle-scroller--item")
print(len(job_cards))
```

```python
cnt = 0
for job_card in job_cards:
    job_title = driver.find_element(By.CSS_SELECTOR, '[data-gtm-joblist="職缺-職缺名稱"]').text
    company = driver.find_element(By.CSS_SELECTOR, '[data-gtm-joblist="職缺-公司名稱"]').text
    industry = driver.find_element(By.CSS_SELECTOR,'[data-gtm-joblist^="職缺-職缺名稱"]').get_attribute('href')
    link = driver.find_element(By.CSS_SELECTOR, '[data-gtm-joblist^="職缺-產業"]').text
    job_desc = driver.find_element(By.CSS_SELECTOR,'.info-description').text
    print(job_title)
    print(company)
    print(industry)
    print(link)
    print(job_desc)
    
```
