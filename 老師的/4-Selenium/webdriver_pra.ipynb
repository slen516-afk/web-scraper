```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
# from webdriver_manager.chrome import ChromeDriverManager

# 建立 Service 物件，指定 chromedriver.exe 的路徑
service = Service('./chromedriver.exe')

# 設定 Chrome 瀏覽器的選項
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized") # Chrome 瀏覽器在啟動時最大化視窗
options.add_argument("--incognito") # 無痕模式
options.add_argument("--disable-popup-blocking") # 停用 Chrome 的彈窗阻擋功能。

# 建立 Chrome 瀏覽器物件
driver = webdriver.Chrome(service=service, options=options)
driver.get("https://www.104.com.tw/")
```

```python
# driver.execute_script("alert('XD')")
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
```

```python
# 寫法 1 - 使用 CLASS_NAME
driver.find_element(By.CLASS_NAME, "form-control").send_keys("Python")
```

```python
# 寫法 2 - 使用 CSS_SELECTOR 找父層後再找子層
cancel_ele = driver.find_element(By.CSS_SELECTOR, ".input-group.input-group--search i")

if cancel_ele:
    try:
        cancel_ele.click()
    except Exception as e:
        print("click 無法點擊")
driver.find_element(By.CSS_SELECTOR, ".input-group.input-group--search input").send_keys("Python")
```

```python
# 寫法 3 - 使用 CSS_SELECTOR 使用屬性精準匹配
driver.find_element(By.CSS_SELECTOR, "input[data-gtm-index='搜尋欄位-搜尋點擊']").send_keys("Python")
```

### 查找不同地區的資料

```python
import time
# 點擊地區選單
driver.find_element(By.CSS_SELECTOR, "button[data-gtm-index='搜尋欄位-搜尋點擊']").click()
time.sleep(2)
# 取得所有區域選項的 value 以提供後續訪問各地區使用

""" 
<a class="category-picker-checkbox">
    <span class="category-picker-checkbox-input">
        <input type="checkbox" class="checkbox-input" value="6001001000">
    </span>
    <span class="children">台北市</span>
</a>
"""
area_elements = driver.find_elements(By.CSS_SELECTOR, ".category-picker__second-floor .second-floor-rect .category-picker-checkbox")
```

```python
print(area_elements[0].find_element(By.CSS_SELECTOR, "input").get_attribute("value"))
print(area_elements[0].find_element(By.CSS_SELECTOR, ".children").text)
```

```python
district_value = {}
for area_element in area_elements:
    area_value = area_element.find_element(By.CSS_SELECTOR, "input").get_attribute("value")
    area_name = area_element.find_element(By.CSS_SELECTOR, ".children").text
    # html = area_element.find_element(By.CSS_SELECTOR, ".children").get_attribute('outerHTML')
    district_value[area_name] = area_value
```

```python
# 確認抓取到的
district_value
```

```python
# 取得台北市的 value 
area = district_value["台北市"] + ',' + district_value['新北市']

# 造訪台北市的搜尋結果頁面
driver.get(f"https://www.104.com.tw/jobs/search/?area={area}&jobsource=index_s&keyword=python&mode=s&page=1")
```

```python
job_cards = driver.find_elements(By.CSS_SELECTOR, ".vue-recycle-scroller__item-view.recycle-scroller--item")
print(len(job_cards))
```

```python
# 取得所有工作項目元素

cnt = 0
for job_card in job_cards:
    job_title = job_card.find_element(By.CSS_SELECTOR, '[data-gtm-joblist="職缺-職缺名稱"]').text
    job_url = job_card.find_element(By.CSS_SELECTOR, '[data-gtm-joblist="職缺-職缺名稱"]').get_attribute('href')
    company = job_card.find_element(By.CSS_SELECTOR, '[data-gtm-joblist="職缺-公司名稱"]').text
    industry = job_card.find_element(By.CSS_SELECTOR, '[data-gtm-joblist^="職缺-產業"]').text
    job_desc = job_card.find_element(By.CSS_SELECTOR, '.info-description').text
    print(job_title)
    print(job_url)
    print(company)
    print(industry)
    print(job_desc)
    cnt += 1
    if cnt >= 3:
        break
```
