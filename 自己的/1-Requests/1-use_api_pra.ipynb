```python
# 引入 requests 模組
import requests

# 測試用 API 網址
url = 'https://httpbin.org/'

# 發送 GET 請求並取得回應
response = requests.get(url)

if response.status_code != 200:
    print("回傳異常")

print(response)
# 回應的狀態碼
print(response.status_code)

# 回應的原始內容（位元組）
print(response.content)
# 回應的文字內容
# print(response.text)# 常用，回傳網頁的html

print(response.headers)
# request = 訪問網頁
# 訪問網頁就會產生header
```

```python
# 使用 GET 搭配參數
r = requests.get(url+'get', params={"city":"Taipei"})
print(r.text)
print(r.json())
data = r.json()
data["args"]
# r(response)轉成json格式的資料

# 等同於使用下面的寫法
r = requests.get(url+'get'+'?city=Taipei')
print(r)
print(r.json())

# 不同程式語言溝通是利用JSON格式
```

```python
# 使用 POST 搭配 data (常用在表單提交)
r = requests.post(url+'post', data={"username":"Ryan"})
print(r.json())
```

```python
# 使用 POST 搭配 JSON (常用在 API 資料傳輸)
r = requests.post(url+'post', json={"username":"Ryan"})
print(r.json())
```
