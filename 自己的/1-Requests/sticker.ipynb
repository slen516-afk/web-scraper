```python
import os

folder_name = "line_stickers"


if not os.path.exists(folder_name):    
    os.mkdir("line_stickers") #make directory
else:
    print(f"已建立過{folder_name}")
```

```python
import json
import requests
from bs4 import BeautifulSoup as bs
    
url = f"https://store.line.me/stickershop/product/30035/zh-Hant"
def get_stickers(url):
    r = requests.get(url)

    soup = bs(r.text, 'lxml') 

    sticker_name = soup.find("p",class_="mdCMN38Item01Ttl")

    #找到貼圖清單
    sticker_list = soup.find("ul", class_="mdCMN09Ul FnStickerList")
    #找到每個貼圖的元素
    sticker_elements = sticker_list.find_all('li')

    #訪問每個貼圖的元素
    for index, sticker_element in enumerate(sticker_elements, start = 1):

        data_preview = json.loads(sticker_element["data-preview"])
        static_url = data_preview["staticUrl"]

        image_content = requests.get(static_url).content #取得二進位圖片內容
        with open(f"line_stickers/line_sticker_{index}.png", "wb") as f:

            f.write(image_content)

    
```
